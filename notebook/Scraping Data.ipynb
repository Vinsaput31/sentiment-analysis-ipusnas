#Data Scraping Review Aplikasi iPunas di Google Play Store

```python
#install module
!pip install google-play-scraper

#import module google_play_scraper untuk scraping data dari google play store
from google_play_scraper import app, reviews, Sort, reviews_all

#import pandas untuk pembuatan dataFrame
import pandas as pd
```

```python
#Scrapping data
scrapreview = reviews_all(
    'mam.reader.ipusnas',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000
)
```

```python
#Menyimpan data ke format csv
df = pd.DataFrame(scrapreview)
df.to_csv('/content/drive/MyDrive/Proyek 1_Dicoding/review_Ipunas.csv')
```
